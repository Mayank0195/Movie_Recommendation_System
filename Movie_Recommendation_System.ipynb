Convert
[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
to
["Action","Adventure","Fantasy","Science Fiction"]

Here as we can see its is a string of list. So we haev to convert these these string of list into list for the proper analysis of the model

```python
import ast
```

```python
#ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

ast is a module in python for this very purpose. So we have imported the required module in the system

```python
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
new_movies['genres'] = new_movies['genres'].apply(convert)
new_movies['genres']
```

```python
new_movies
```

```python
def transform(obj):
  M = []
  for i in ast.literal_eval(obj):
    M.append(i['name'])
  return M
```

```python
new_movies['keywords'] = new_movies['keywords'].apply(transform)
```

```python
new_movies
```

```python
def convert3(obj):
  m = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      m.append(i['name'])
      counter +=1
    else:
      break
  return m     

```

```python
new_movies['cast'] = new_movies['cast'].apply(convert3)
```

```python
new_movies.head()
```

```python
def fetch_director(obj):
  p = []
  for i in ast.literal_eval(obj):
      if i['job'] == 'Director':
        p.append(i['name'])
  return p   
```

```python
new_movies['crew'] = new_movies['crew'].apply(fetch_director)
```

```python
new_movies.head()
```

```python
new_movies['overview'][0]
```

```python
new_movies['overview'] = new_movies['overview'].apply(lambda x:x.split())
```

```python
new_movies['genres'] = new_movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
new_movies['keywords'] = new_movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
new_movies['cast'] = new_movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
new_movies['crew'] = new_movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
new_movies
```

```python
new_movies['tags'] = new_movies['overview'] + new_movies['genres'] + new_movies['keywords'] + new_movies['cast'] + new_movies['crew']
```

```python
new_df = new_movies[['movie_id','title','tags']]
new_df
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)  
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 7000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray( )
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
 cosine_similarity(vectors).shape
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]  
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
    

```

```python
recommend('Batman Begins')
```
